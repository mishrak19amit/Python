```python
Topfiftycandidate=df.iloc[:50,:]
```

```python
import matplotlib
Topfiftycandidate
```

```python
import matplotlib.pyplot as plt
```

```python
x = Topfiftycandidate['RANK.'] 
# corresponding y axis values 
y = Topfiftycandidate['MAIN_MARKS'] 

plt.plot(x, y, color='blue', linestyle='dashed', linewidth = 1, 
         marker='o', markerfacecolor='red', markersize=5) 
  
# naming the x axis 
plt.xlabel('x - Ranks') 
# naming the y axis 
plt.ylabel('y - Mains Marks') 
  
# giving a title to my graph 
plt.title('Mains_Marks_V/S_Rank graph!') 
  
# function to show the plot 
plt.show() 
```

```python
x = Topfiftycandidate['RANK.'] 
# corresponding y axis values 
y = Topfiftycandidate['PT_MARKS'] 

plt.plot(x, y, color='green', linestyle='dashed', linewidth = 1, 
         marker='o', markerfacecolor='blue', markersize=5) 
  
# naming the x axis 
plt.xlabel('x - Ranks') 
# naming the y axis 
plt.ylabel('y - Mains Marks') 
  
# giving a title to my graph 
plt.title('PT_Marks_V/S_Rank graph!') 
  
# function to show the plot 
plt.show() 
```

```python
#ST=61, OBC=209, SC=128, GEN=361, TOTAL=759

Cat=["GEN", "OBC", "SC", "ST"]
SelectionCatWise=[361, 209, 128, 61]
```

```python
dfcatselection = pd.DataFrame(list(zip(Cat, SelectionCatWise)), 
               columns =['Category', 'Selection']) 
dfcatselection
```

```python
labels = 'GEN', 'OBC', 'SC', 'ST'
explode = (0.05, 0.04, 0.03, 0.03)
fig1, ax1 = plt.subplots()
ax1.pie(SelectionCatWise, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```

```python
plt.bar(dfcatselection['Category'], dfcatselection['Selection'], ) 
plt.xlabel("Category") 
plt.ylabel("Selection") 
plt.show() 
```
